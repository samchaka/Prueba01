```python
import pandas as pd
import numpy as np
import glob
import os
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
```

```python
all_files = glob.glob("/content/drive/MyDrive/ArchivosIncidencias_20212023/*xlsx")
len(all_files)
```

```python
file_list = []
```

```python
for f in all_files:
    data = pd.read_excel(f,sheet_name = 1,
                     skiprows = 6,
                     skipfooter = 0,
                     header = None,
                     usecols = "A,B,C,F,H,J,L,P,AB,AC,AE,AI,AK,AO,AW,AX,AY,AZ,BB,BC,BD",
                     names = ["contrato","no_empleado","nombre","prima_dominical","festivo","descanso_trabajado","horas_dobles","suspensiones","transporte","faltas","PSGS","incapacidad","maternidad","vacaciones","horas_triples","suplencias","doble_turno","observaciones","alimentos","inicio_qna", "qna"])

    archivo = os.path.basename(f)
    data ["archivo"] = archivo #nueva columan con el nombre del archivo(archivo)
    file_list.append(data)
```

```python
df = pd.concat(file_list,ignore_index = True)
```

```python
df_puesto = pd.read_excel("/content/drive/MyDrive/Sindicalizados_puesto.xlsx", usecols= "A,C")
```

```python
df_puesto.set_index("no_empleado", inplace=True)
```

```python
df = pd.merge(df,df_puesto,how="left", on="no_empleado")
```

```python
df.index = df['no_empleado'] #Toma No_empleado como indice.
df.drop(["no_empleado"],axis = 1, inplace = True) #Elimina la columna No_empleado.
```

```python
df.sort_values(by="inicio_qna",ascending = True, inplace = True)
```

```python
df["inicio_qna"] = pd.to_datetime(df["inicio_qna"]) #define la columna Inicio_qna como datos de tiempo.
```
